```python
import pandas as pd
path = "../data/raw/Electronics_5.json"
df = pd.read_json(path, lines=True)
print(df.shape)
df.head()
```

```python
cols = [
    "reviewerID",
    "asin",
    "overall",
    "reviewText",
    "unixReviewTime"
]
df = df[cols]
```

```python
df = df.dropna(subset=["reviewerID", "asin", "overall"])
df = df[df["overall"].between(1, 5)]
df["reviewText"] = df["reviewText"].fillna("")
```

```python
user_counts = df["reviewerID"].value_counts()
item_counts = df["asin"].value_counts()
valid_users = user_counts[(user_counts >= 5) & (user_counts <= 50)].index
valid_items = item_counts[(item_counts >= 5) & (item_counts <= 100)].index
df = df[
    df["reviewerID"].isin(valid_users) &
    df["asin"].isin(valid_items)
]
```

```python
df = df.sample(n=500_000, random_state=42)
print(df.shape)
```

```python
import matplotlib.pyplot as plt
df["overall"].value_counts().sort_index().plot(kind="bar")
plt.title("Rating Distribution")
plt.show()
```
